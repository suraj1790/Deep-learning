```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("brest_cancer_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(columns=["Unnamed: 32","id"],inplace=True)
```

```python
df.info()
```

```python
df["diagnosis"].value_counts()
```

```python
df.describe()
```

```python
val = {"M":0,"B":1}
df["diagnosis"] = df["diagnosis"].map(val)
```

```python
df.info()
```

```python
x = df.drop(columns="diagnosis")
y = df["diagnosis"].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# scaled.val
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
# build newral network
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
```

```python
model = Sequential()
model.add(Flatten(input_shape=(30,)))
```

```python
model.add(Dense(10,activation="relu"))
model.add(Dense(56,activation="relu"))
model.add(Dense(5,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
hist = model.fit(x_train_scaled,y_train,validation_split=0.2,epochs=50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history["accuracy"])
plt.plot(hist.history["val_accuracy"])
plt.legend(["training data","validation data"],loc = 'lower right')
```

```python
plt.plot(hist.history["loss"])
plt.plot(hist.history["val_loss"])

plt.legend(["traing data","validation data"])
```

```python
loss,accuracy = model.evaluate(x_test_scaled,y_test)
```

```python
loss,accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_prob = model.predict(x_test_scaled)
y_pred = y_prob.argmax(axis=1)
print(y_pred)
accuracy_score(y_test,y_pred)
```
